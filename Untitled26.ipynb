```python
df = pd.read_csv('https://raw.githubusercontent.com/issabellasafitri/SuperApp/main/Data-Olah-Super-App.csv')
df
```

```python
df[['pendapatan_min', 'pendapatan_max']] = df['Pendapatan'].str.split('-', expand=True)

# gua buang kolom pendapatan karena udah ga kepake
df = df.drop(['Pendapatan'], axis=1)

# gua ubah isi dari kolom "Gen"
df['Gen'] = df['Gen'].replace({0:'X', 1:'Y', 2:'Z'})
```

```python
df
```

```python
df['Gen'] = df['Gen'] .astype('category')
df['Domisili'] = df['Domisili'] .astype('category')
df['Status'] = df['Status'] .astype('category')
df['pendapatan_min'] = df['pendapatan_min'] .astype('int')
df['pendapatan_max'] = df['pendapatan_max'] .astype('int')
df['Pekerjaan'] = df['Pekerjaan'] .astype('category')
df['App_Perbankan'] = df['App_Perbankan'] .astype('category')
df['Skala'] = df['Skala'] .astype('category')
df['Fitur'] = df['Fitur'] .astype('category')
df['Permasalahan'] = df['Permasalahan'] .astype('category')
```

```python
import plotly.express as px
import dash
from dash import dcc, html, Input, Output
from jupyter_dash import JupyterDash
```

```python
df_bar = df.groupby(['App_Perbankan', 'Gen']).size().reset_index(name='Jumlah')

fig = px.bar(df_bar, x="App_Perbankan", y= "Jumlah", color="Gen", barmode='group', labels={"App_Perbankan":"Nama Aplikasi Bank"}, 
             title='Aplikasi Perbankan yang Dipakai Masing-Masing Generasi')
fig.show()
```

```python
app = JupyterDash(__name__)

app.layout = html.Div([
    html.H4('Fitur yang digunakan dari masing-masing generasi'),
    dcc.Dropdown(id='dropdown',
                 options=["Transfer Uang", 
                          "Cek Saldo dan Mutasi Rekening", 
                          "Pembayaran atau Pembelian",
                          "Top Up Ewallet", 
                          "Pembayaran menggunakan QR",
                          "Tarik uang"],
                 value='Tarik uang'
                ),
    dcc.Graph(id='graph')
])

@app.callback(
    Output("graph", "figure"), 
    Input("dropdown", "value"))

def update_figure(value):
    
    df_baru = df[df['Fitur'].str.contains(value)]
    df_baru = df_baru['Gen'].value_counts().rename("jumlah").reset_index()
    
    fig = px.bar(df_baru, x="index", y="jumlah")
    return fig

if __name__ == '__main__':
    app.run_server(mode="inline")
```
